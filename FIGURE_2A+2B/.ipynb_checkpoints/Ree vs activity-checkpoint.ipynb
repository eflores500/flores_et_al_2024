```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import seaborn as sns
import glob
import warnings
warnings.filterwarnings('ignore')


sns.set(rc={"figure.dpi":1200})
sns.set_context('notebook')
sns.set_style("ticks")
mpl.style.use('default')
mpl.rcParams['axes.linewidth'] = 7 #set the value globally
mpl.rcParams['xtick.major.size'] = 20
mpl.rcParams['xtick.major.width'] = 7
mpl.rcParams['xtick.minor.size'] = 10
mpl.rcParams['xtick.minor.width'] = 7
mpl.rcParams['ytick.major.size'] = 20
mpl.rcParams['ytick.major.width'] = 7
mpl.rcParams['ytick.labelsize'] = 50
mpl.rcParams['xtick.labelsize'] = 50
mpl.rcParams['ytick.minor.size'] = 10
mpl.rcParams['ytick.minor.width'] = 7
mpl.rcParams['font.size'] = 55
mpl.rcParams['font.sans-serif']='Arial'
```

```python
# Replace 'file_path.csv' with your file's path
df = pd.read_csv('delrosso_AD_ONLY_complete_albatross_wACTIVITY.csv')
GS = pd.read_csv('delrosso_GS_albatross_DATA.csv')
# Merge df with GS on the matching values of 'Length' in df and 'GS_length' in GS
# 'how' parameter is set to 'left' to keep all rows from df and only matching rows from GS
# 'left_on' and 'right_on' parameters specify the columns to merge on in df and GS respectively
merged_df = pd.merge(df, GS, how='left', left_on='Length', right_on='GS_length')

# The merged dataframe will have all columns from df and GS, including 'Re' from GS
# You can then create a new column 'Re_GS' which is a copy of 'Re' column from GS
df['Re_GS'] = merged_df['Re']
df['Re_normalized'] = df['Ree (nm)'] / df['Re_GS']


# Calculate the minimum and maximum values of the 'Activity' column
min_value = df['Max avg act'].min()
max_value = df['Max avg act'].max()

# Apply the min-max normalization
df['Activity_Normalized'] = (df['Max avg act'] - min_value) / (max_value - min_value)
```

```python
# Selecting only the 'Re_GS' and 'Length' columns from df
selected_columns = df[['Re_GS', 'Length']]

# Dropping duplicate rows based on all columns in selected_columns
unique_values = selected_columns.drop_duplicates()
sorted_unique_values = unique_values.sort_values(by='Length')

# Display the unique values
print(sorted_unique_values)
```

```python
# Sample data for demonstration
# In actual use, replace this with your df['Re_normalized'] and df['Activity_Normalized']
re_normalized = df['Re_normalized']
activity_normalized = df['Activity_Normalized']

# Creating the plot with square dimensions
plt.figure(figsize=(6, 6))
plt.scatter(re_normalized, activity_normalized)
plt.xlabel('Re normalized',fontsize=30)
plt.ylabel('Normalized Activity',fontsize=30)
plt.title('Activity vs Re normalized',fontsize=30)
# Adding an additional tick on the x-axis

current_ticks = plt.xticks()[0]  # Get current x-axis ticks
new_tick = (max(current_ticks) - min(current_ticks)) / 2  # Calculate new tick position
new_ticks = sorted(list(current_ticks) + [new_tick])  # Add new tick to the list
plt.xticks(new_ticks, fontsize = 30)  # Set the new ticks
plt.xlim(0.75, 1.3)
plt.show()

```
