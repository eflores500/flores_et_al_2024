```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import seaborn as sns
import glob
import warnings
warnings.filterwarnings('ignore')
sns.set(rc={"figure.dpi":1200})
sns.set_context('notebook')
sns.set_style("ticks")
mpl.style.use('default')
mpl.rcParams['axes.linewidth'] = 7 #set the value globally
mpl.rcParams['xtick.major.size'] = 20
mpl.rcParams['xtick.major.width'] = 7
mpl.rcParams['xtick.minor.size'] = 10
mpl.rcParams['xtick.minor.width'] = 7
mpl.rcParams['ytick.major.size'] = 20
mpl.rcParams['ytick.major.width'] = 7
mpl.rcParams['ytick.labelsize'] = 50
mpl.rcParams['xtick.labelsize'] = 50
mpl.rcParams['ytick.minor.size'] = 10
mpl.rcParams['ytick.minor.width'] = 7
mpl.rcParams['font.size'] = 50
mpl.rcParams['font.sans-serif']='Arial'
```

```python
activity = pd.read_csv('Shahar_ADvariants_20211203.csv')
activity = activity[activity['ID'].isin([1, 2, 3, 4, 5])]
activity = activity.drop(columns=['norm_error','start','end'])
activity= activity.rename(columns={'MutName': 'mut name'})
activity['Activity_Mean_MSS18'] = activity['Activity_Mean_MSS18'] - 2000
activity['mut name'] = activity['mut name'].replace('supercharge6', 'RK>E')

wt_value = activity.loc[activity['mut name'] == 'WT', 'Activity_Mean_MSS18'].iloc[0]

# Create a new column 'normalized activity' by dividing 'Activity_Mean_MSS18' by the WT value
activity['normalized activity'] = activity['Activity_Mean_MSS18'] / wt_value
# Find the standard deviation value for the 'WT' mut name
wt_std = activity[activity['mut name'] == 'WT']['Activity_std_MSS18'].iloc[0]

# Calculate the normalized error
activity['normalized error'] = ((activity['Activity_std_MSS18'] / activity['Activity_Mean_MSS18'])**2 + 
                                (wt_std / wt_value)**2)**0.5 * activity['normalized activity']

activity
```

```python
import matplotlib.pyplot as plt

# Ensure your DataFrame 'activity' is properly loaded with the necessary columns

# Swapping the values of 'MA>W' and 'NQ>W' in the 'mut name' column as well as their corresponding values
ma_w_index = activity[activity['mut name'] == 'MA>W'].index.tolist()
nq_w_index = activity[activity['mut name'] == 'NQ>W'].index.tolist()

# Check if both indices are found
if ma_w_index and nq_w_index:
    ma_w_index, nq_w_index = ma_w_index[0], nq_w_index[0]
    
    # Swap the 'mut name' values
    activity.at[ma_w_index, 'mut name'], activity.at[nq_w_index, 'mut name'] = activity.at[nq_w_index, 'mut name'], activity.at[ma_w_index, 'mut name']
    
    # Swap the 'normalized activity' values
    activity.at[ma_w_index, 'normalized activity'], activity.at[nq_w_index, 'normalized activity'] = activity.at[nq_w_index, 'normalized activity'], activity.at[ma_w_index, 'normalized activity']

    # Swap the 'normalized error' values if present
    if 'normalized error' in activity.columns:
        activity.at[ma_w_index, 'normalized error'], activity.at[nq_w_index, 'normalized error'] = activity.at[nq_w_index, 'normalized error'], activity.at[ma_w_index, 'normalized error']

# Creating a scatter plot with the updated DataFrame
plt.figure(figsize=(10, 6))
plt.scatter(activity['mut name'], activity['normalized activity'], color='magenta', s=200)  # Scatter plot
plt.errorbar(activity['mut name'], activity['normalized activity'], yerr=activity['normalized error'], fmt='none', ecolor='magenta', elinewidth=2.5, capsize=5)  # Adding error bars
plt.xlabel('Mutant', fontsize=40)
plt.ylabel('Normalized Activity', fontsize=40)
plt.xticks(fontsize=28, rotation=45)  # Rotating x-axis labels for better visibility

plt.show()
```
