```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import seaborn as sns
import glob
import warnings
warnings.filterwarnings('ignore')
```

```python
###IMPORT DATAFRAME###
cells = pd.read_csv('MASTER_HIF1A_CITED2_HEK_U2OS_220920.csv',index_col=0)

##FILTERS WITHOUT SIZE##
cells=cells[(cells['D/A_corr']>0) & (cells['D/A_corr']<3)]
cells=cells[(cells['directA']<30000)]
cells=cells[cells['Cell Line'].str.contains("U2OS")]
##DIVIDE, SELECT SIZE, REJOIN DF
# HEK=cells[cells['Cell Line'].str.contains("HEK")]
# HEK=HEK[(HEK['area']<1000)]
# U2OS=cells[cells['Cell Line'].str.contains("U2OS")]
# U2OS=U2OS[(U2OS['area']<1300)]
# rejoin=[HEK,U2OS]
# cells = pd.concat(rejoin)
# cells=cells[(cells['ID']!=6)]
# cells['log directA'] = np.log10(cells['directA'])




```

```python
##DEFINE HIF DATAFRAME###
HIF=cells[(cells['ID']<6)]
sns.violinplot(x=HIF['ID'],y=HIF['Ef'], hue = HIF['Cell Line'])
```

```python
###IMPORT ACTIVITY###
TA_HIF = pd.read_csv('Shahar_ADvariants_20211203.csv')
TA_HIF = TA_HIF[TA_HIF['ID'].isin([ 1, 2, 3, 4,5 ])].sort_values(by=['ID'])
TA_HIF_norm = TA_HIF[TA_HIF['MutName']=='WT']['Activity_Mean_MSS18'].values

TA_HIF['Activity_normalized'] = TA_HIF['Activity_Mean_MSS18']/TA_HIF_norm
#TA_HIF
```

```python
TA_HIF
```

```python


# Sample data (replace this with your actual data or data loading code)
# TA_HIF = pd.DataFrame({
#     'Activity_Mean_MSS18': [250, 180, 300, 100, 230]
# })

# Assuming you've already loaded your dataframe as TA_HIF

TA_HIF['Activity_Mean_MSS18'] = TA_HIF['Activity_Mean_MSS18'].apply(lambda x: x - 2000 if x > 200 else 0)

TA_HIF
```

```python

# Assuming you've loaded your dataframe as TA_HIF

# Extract 'Activity_Mean_MSS18' value for 'WT'
wt_value = TA_HIF[TA_HIF['MutName'] == 'WT']['Activity_Mean_MSS18'].iloc[0]
wt_std = TA_HIF[TA_HIF['MutName'] == 'WT']['Activity_std_MSS18'].iloc[0]

# Normalize 'Activity_Mean_MSS18' by dividing every value by the WT value
TA_HIF['Normalized_Activity_Mean_MSS18'] = TA_HIF['Activity_Mean_MSS18'] / wt_value

# Calculate the error propagation for the normalized value
TA_HIF['Normalized_Error'] = TA_HIF['Normalized_Activity_Mean_MSS18'] * (
    ((TA_HIF['Activity_std_MSS18'] / TA_HIF['Activity_Mean_MSS18'])**2 +
     (wt_std / wt_value)**2)**0.5
)

TA_HIF
```

```python
##HIF1A_ACTIVITY
sns.reset_defaults()
sns.set(rc={"figure.dpi":600})
sns.set_context('notebook')
sns.set_style("ticks")

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax2.set_ylim(0, 2.2)
ax1.set_ylim(0.2, 0.8)
plt.grid(alpha=0.5)

 

sns.scatterplot(x=TA_HIF['ID']-1,y=TA_HIF['Activity_normalized'],s=100,palette='tab10',color='r')
sns.violinplot(x=HIF['ID'],y=HIF['Ef'],size=4,ax=ax1, color = 'deepskyblue')
plt.errorbar(x=TA_HIF['ID']-1,y=TA_HIF['Activity_normalized'],yerr=TA_HIF['norm_error'],fmt='o',color='magenta')

plt.xticks([0,1,2,3,4],['NQ>W','MA>W','WT','supercharge6','RK>D'])
ax1.set(xlabel='Sequence')

plt.title('HIF1\u03B1')
ax2.yaxis.label.set_color('magenta')
ax2.spines['right'].set_color('magenta')
ax2.tick_params(axis='y', colors='magenta')

#ax1.legend(bbox_to_anchor=(1.15, 1), loc='upper left', borderaxespad=0)
ax2.set(ylabel='Normalized Activity')
ax1.set(ylabel='E$^{Cell}_{f}$')
plt.show()



#fig.savefig('Hif1a.png')


```
