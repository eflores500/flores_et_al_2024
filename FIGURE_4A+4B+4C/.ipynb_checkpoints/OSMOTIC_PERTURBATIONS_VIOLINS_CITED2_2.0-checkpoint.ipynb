```python
# Find unique IDR values and the number of unique values
unique_idrs = df['IDR'].unique()
n_unique_idrs = len(unique_idrs)

# Create a figure with subplots - adjust the number of rows/columns as needed
# For example, for 4 unique IDR values, you might want a 2x2 grid
n_rows = int(n_unique_idrs ** 0.5) + (1 if n_unique_idrs % int(n_unique_idrs ** 0.5) > 0 else 0)
n_cols = int(n_unique_idrs / n_rows) + (1 if n_unique_idrs % n_rows > 0 else 0)

fig, axs = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 5 * n_rows))

for i, idr in enumerate(unique_idrs):
    # Filter df for the current IDR value
    df_filtered = df[df['IDR'] == idr]
    
    # Determine the current subplot
    ax = axs[i // n_cols, i % n_cols] if n_unique_idrs > 1 else axs
    
    # Plotting
    sns.scatterplot(data=df_filtered, x='log directA_before', y='Ef_after', ax=ax)
    ax.set_title(f'IDR: {idr}')

# Adjust layout for better readability
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Calculate median 'delta Ef' for each 'replicate' of each 'condition' for each 'mut name'
df_median = df.groupby(['mut name', 'condition', 'well'])['delta Ef'].median().reset_index()

# Rename 'delta Ef' column to 'median delta Ef'
df_median.rename(columns={'delta Ef': 'median delta Ef'}, inplace=True)

# Calculate standard deviation of 'median delta Ef' for each 'condition' for each 'mut name'
df_std = df_median.groupby(['mut name', 'condition'])['median delta Ef'].std().reset_index()

# Rename 'median delta Ef' column to 'std dev delta Ef'
df_std.rename(columns={'median delta Ef': 'std dev delta Ef'}, inplace=True)

# Merge df_median and df_std
df_final = pd.merge(df_median, df_std, on=['mut name', 'condition'], how='left')


# Calculate average 'median delta Ef' for each 'condition' for each 'mut name'
df_mean = df_final.groupby(['mut name', 'condition'])['median delta Ef'].mean().reset_index()

# Rename 'median delta Ef' column to 'average median delta Ef'
df_mean.rename(columns={'median delta Ef': 'average median delta Ef'}, inplace=True)

# Calculate standard deviation of 'average median delta Ef' for each 'condition' for each 'mut name'
df_std = df_final.groupby(['mut name', 'condition'])['median delta Ef'].std().reset_index()

# Rename 'median delta Ef' column to 'std dev average median delta Ef'
df_std.rename(columns={'median delta Ef': 'std dev average median delta Ef'}, inplace=True)

# Merge df_mean and df_std
df_final_avg = pd.merge(df_mean, df_std, on=['mut name', 'condition'], how='left')

# Print the final DataFrame
print(df_final_avg)
```

```python
# Filter df_final_avg for 'condition' 750 and 100
df_750 = df_final_avg[df_final_avg['condition'] == 750]
df_100 = df_final_avg[df_final_avg['condition'] == 100]

# Set 'mut name' as the index for easy subtraction
df_750.set_index('mut name', inplace=True)
df_100.set_index('mut name', inplace=True)

# Calculate the absolute difference in 'average median delta Ef' between 'condition' 750 and 100 for each 'mut name'
df_diff = (df_750['average median delta Ef'] - df_100['average median delta Ef']).abs()

# Calculate the propagated standard error
df_error = (df_750['std dev average median delta Ef']**2 + df_100['std dev average median delta Ef']**2)**0.5

# Create the sensitivity DataFrame
sensitivity = pd.DataFrame({'absolute difference': df_diff, 'propagated_SE': df_error})

# Reset the index
sensitivity.reset_index(inplace=True)

# Print the sensitivity DataFrame
print(sensitivity)
```

```python
import scipy.stats as stats

# Define the desired order for 'mut name'
mut_name_order = ['FML>A', 'L>F', 'L>W', 'WT', 'RK>E', 'RK>Amut20to39', 'GS24']

# Filter the dataframe for the hypo and hyper conditions
hypo_conditions = [100, 300]
hyper_conditions = [300, 750]

hypo = df[df['condition'].isin(hypo_conditions)]
hyper = df[df['condition'].isin(hyper_conditions)]

condition_colors = {100: 'green', 300: 'orange', 750: 'blue'}

def compute_p_value(mutant, df_subset):
    conditions = df_subset['condition'].unique()
    
    if len(conditions) != 2:
        raise ValueError("The dataframe should only contain two unique conditions for comparison.")

    condition_1_data = df_median[(df_median['mut name'] == mutant) & (df_median['condition'] == conditions[0])]['median delta Ef']
    condition_2_data = df_median[(df_median['mut name'] == mutant) & (df_median['condition'] == conditions[1])]['median delta Ef']
    
    min_length = min(len(condition_1_data), len(condition_2_data))
    
    _, p_val = stats.ttest_rel(condition_1_data[:min_length], condition_2_data[:min_length])
    
    return p_val
def is_significant(mutant, df_subset):
    p_val = compute_p_value(mutant, df_subset)
    if p_val < 0.0001:
        return '***'  # Triple asterisk for p < 0.0001
    elif p_val < 0.001:
        return '**'   # Double asterisk for p < 0.001
    elif p_val < 0.01:
        return '*'    # Singleaasterisk for p < 0.01
    else:
        return ''     # No asterisk for p >= 0.01

def plot_data(df_subset, filename):
    conditions = df_subset['condition'].unique()
    wells = df_subset['well'].unique()

    fig, axes = plt.subplots(nrows=1, ncols=len(mut_name_order), figsize=(20, 4), sharey=True)

    for idx, mut_name in enumerate(mut_name_order):
        ax = axes[idx]
        ax.set_title(mut_name, fontsize=20)

        for j, condition in enumerate(conditions):
            for k, well in enumerate(wells):
                filtered_data = df_subset[(df_subset['mut name'] == mut_name) & 
                                          (df_subset['condition'] == condition) & 
                                          (df_subset['well'] == well)]['delta Ef']

                if not filtered_data.empty:
                    vp = ax.violinplot(filtered_data, positions=[j], widths=0.7,
                                       showmeans=False, showextrema=False, showmedians=False)

                    for pc in vp['bodies']:
                        pc.set_facecolor('white')
                        pc.set_edgecolor(condition_colors[condition])
                        pc.set_alpha(0.45)

                    avg_median_delta_Ef = df_final_avg.loc[(df_final_avg['mut name'] == mut_name) & 
                                                           (df_final_avg['condition'] == condition), 
                                                           'average median delta Ef'].values[0]
                    std_dev_avg_median_delta_Ef = df_final_avg.loc[(df_final_avg['mut name'] == mut_name) & 
                                                                   (df_final_avg['condition'] == condition), 
                                                                   'std dev average median delta Ef'].values[0]

                    if not pd.isna(std_dev_avg_median_delta_Ef):
                        ax.errorbar(j, avg_median_delta_Ef, 
                                    yerr=std_dev_avg_median_delta_Ef, fmt='o', color='red', zorder=2)

                    # Check for statistical significance
                    significance = is_significant(mut_name, df_subset)
                    if significance:
                        ax.annotate(significance, (len(conditions) / 2 - 0.5, 0.085), fontsize=30, ha='center', va='center', color='black')

        ax.set_xticks(range(len(conditions)))
        ax.set_xticklabels(conditions, rotation=360, fontsize=30)
        
        
        # Adding gray dashed line in the middle
        ax.axhline(0, color='gray', linestyle='--')
        
    axes[0].set_ylabel('$\Delta$E$_{{f}}$',fontsize=30)
    legend_artists = [plt.Line2D([0], [0], color=color, lw=4) for color in [condition_colors[c] for c in conditions]]
    legend= axes[-1].legend(legend_artists, [str(condition) + ' mOsm' for condition in conditions],fontsize=15, bbox_to_anchor=(2,1))
        # Set the font size for each text element in the legend
#     for text in legend.get_texts():
#         text.set_fontsize(8)
    plt.ylim(-.15,.12)
    #plt.tight_layout()
    plt.savefig("{}.png".format(filename))
    plt.show()

    # Print the p-values for each mut name
    p_values = {mutant: compute_p_value(mutant, df_subset) for mutant in mut_name_order}
    p_values_df = pd.DataFrame(list(p_values.items()), columns=['Mutant', 'P-value'])
    print(p_values_df)

# Call the function for both hypo and hyper conditions
plot_data(hypo, "hypo_results")
plot_data(hyper, "hyper_results")
```

```python
def calculate_counts(df):
    # Group by 'mut name' and 'condition', then count the occurrences
    counts = df.groupby(['mut name', 'well']).size().reset_index(name='Count')
    
    return counts

# Example usage:
counts_df = calculate_counts(df)
print(counts_df)
```

```python
def calculate_counts(df):
    # Group by 'mut name' and count the unique wells
    counts = df.groupby('mut name')['well'].nunique().reset_index(name='Well Count')
    
    return counts

# Example usage:
counts_df = calculate_counts(df)
print(counts_df)
```
