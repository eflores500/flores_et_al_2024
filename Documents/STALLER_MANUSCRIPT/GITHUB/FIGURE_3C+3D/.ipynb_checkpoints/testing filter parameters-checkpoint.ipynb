```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import seaborn as sns
import glob
import warnings


df = pd.read_csv('U2OS_HIF1A_MASTER_with_WELLS.csv',index_col=0)
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import math


# Assuming you have a DataFrame named "df" with columns 'IDR', 'log directA_before', and 'Ef_before'

# Get unique IDRs from the DataFrame
unique_idrs = df['IDR'].unique()

# Calculate the number of rows and columns for the subplot grid
num_rows = math.ceil(len(unique_idrs) / 2)
num_cols = 2

# Create a grid of subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6 * num_rows))

for i, idr in enumerate(unique_idrs):
    # Calculate the row and column indices for the current subplot
    row = i // num_cols
    col = i % num_cols
    
    # Filter the DataFrame for the current IDR
    subset = df[df['IDR'] == idr]
    
    # Create a scatter plot with 'log directA_before' on the x-axis and 'Ef_before' on the y-axis
    axes[row, col].scatter(subset['directA'], subset['Ef_after'])
    
    # Set title for the subplot
    axes[row, col].set_title(f'IDR {idr}')
    
    # Add labels
    axes[row, col].set_xlabel('directA')
    axes[row, col].set_ylabel('Ef_after')
# Set x-axis limits to 10000
    #axes[row, col].set_xlim(0, 2000)
# Remove any empty subplots if there are an odd number of IDRs
if len(unique_idrs) % 2 != 0:
    fig.delaxes(axes[-1, -1])

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Assuming you have a DataFrame named "df" with columns 'IDR', 'd_directA', 'Ef_after', and 'repeat'

# Get unique IDRs from the DataFrame
unique_idrs = df['IDR'].unique()

# Calculate the number of rows and columns for the subplot grid
num_rows = math.ceil(len(unique_idrs) / 2)
num_cols = 2

# Create a grid of subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6 * num_rows), sharex=True, sharey=True)

for i, idr in enumerate(unique_idrs):
    # Calculate the row and column indices for the current subplot
    row = i // num_cols
    col = i % num_cols
    
    # Filter the DataFrame for the current IDR
    subset = df[df['IDR'] == idr]
    
    # Create a histogram with 'd_directA' on the x-axis
    axes[row, col].hist(subset['d_directA'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    
    # Set title for the subplot
    axes[row, col].set_title(f'IDR {idr}')
    
    # Add labels
    axes[row, col].set_xlabel('d_directA')
    axes[row, col].set_ylabel('Count')
    
    # Set x-axis limits to 10000 (if desired)
    # axes[row, col].set_xlim(0, 10000)

# Remove any empty subplots if there are an odd number of IDRs
if len(unique_idrs) % 2 != 0:
    fig.delaxes(axes[-1, -1])

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
##assigning filters 
##REMEMBER CUTOFF IS AT 15000, USING DIRECTA
df_filtered = df[(df['D_before'] < 50000) & (df['A_before'] < 50000)]

df_filtered = df_filtered[(df_filtered['D_after']<50000)]
df_filtered = df_filtered[(df_filtered['A_after'] > 500) & (df_filtered['A_after'] < 50000)]
df_filtered = df_filtered[(df_filtered['Ef_before']>0)]
df_filtered = df_filtered[(df_filtered['Ef_after']>0)]
df_filtered = df_filtered[(df_filtered['area_before']<40000)]
df_filtered = df_filtered[(df_filtered['d_directA']<30000)]


df_filtered
```

```python
# Assuming you have a DataFrame named "df" with columns 'IDR', 'd_directA', 'Ef_after', and 'repeat'


# Get unique IDRs from the filtered DataFrame
unique_idrs = df_filtered['IDR'].unique()

# Calculate the number of rows and columns for the subplot grid
num_rows = math.ceil(len(unique_idrs) / 2)
num_cols = 2

# Create a grid of subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6 * num_rows), sharex=True, sharey=True)

for i, idr in enumerate(unique_idrs):
    # Calculate the row and column indices for the current subplot
    row = i // num_cols
    col = i % num_cols
    
    # Filter the filtered DataFrame for the current IDR
    subset = df_filtered[df_filtered['IDR'] == idr]
    
    # Create a scatter plot with 'd_directA' on the x-axis, 'Ef_after' on the y-axis, and color-coded by 'repeat'
    scatter = axes[row, col].scatter(subset['directA_before'], subset['Ef_before'], c=subset['repeat'], cmap='tab10')
    
    # Set title for the subplot
    axes[row, col].set_title(f'IDR {idr}')
    
    # Add labels
    axes[row, col].set_xlabel('log directA_before')
    axes[row, col].set_ylabel('Ef_before')
    
    # Set x-axis limits to 10000
    #axes[row, col].set_xlim(0, 10000)

# Remove any empty subplots if there are an odd number of IDRs
if len(unique_idrs) % 2 != 0:
    fig.delaxes(axes[-1, -1])

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
