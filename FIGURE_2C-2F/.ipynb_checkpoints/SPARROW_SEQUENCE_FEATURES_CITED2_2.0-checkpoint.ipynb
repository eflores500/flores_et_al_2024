```python
df = pd.read_csv('HIF1A_CITED2_SIGNIFICANT_ACTIVITY_STALLER.csv')
df = df[df['-log10_p-values'] > 1.30102999566]

####REMOVE ANY SEQUENCE THAT DOESN'T HAVE A STANDARD AMINO ACID
# Assuming df is your DataFrame and it already exists with a column 'ADseq' containing the sequences

standard_aa = set('ACDEFGHIKLMNPQRSTVWY')  # Set of standard amino acid characters

# Use DataFrame.apply to check each sequence, keeping only those that are subsets of standard_aa
df = df[df['ADseq'].apply(lambda x: set(x.upper()).issubset(standard_aa))]

# Now 'df' will include a 'status' column with the categories 'activating', 'deactivating', or 'neutral'
# Assuming 'df' is your DataFrame and 'log2_FC' is your column of interest
df['status'] = df['log2_FC'].apply(lambda x: 'deactivating' if x < -1 else ('activating' if x > 1 else 'neutral'))

df
```

```python
print(df.columns)
df
```

```python
import pandas as pd
from sparrow import Protein
# from sparrow import AnalyticalFRC

# Initialize an empty DataFrame for storing results
results_df = pd.DataFrame()

# Iterate over each sequence in the 'ADseq' column of df
for idx, IDR_seq in df['ADseq'].items():  # Use .items() for Series
    # Initialize a dictionary to hold the data for this sequence
    IDR_data = {'ADseq': IDR_seq}  # Add the sequence itself to IDR_data
    
    # Predict properties using Sparrow
    P = Protein(IDR_seq)
    IDR_data['Rg'] = P.predictor.radius_of_gyration()
    IDR_data['Re'] = P.predictor.end_to_end_distance(use_scaled=False)
    IDR_data['aspher'] = P.predictor.asphericity()
    
    # Get other properties of IDR
    IDR_data['NCPR'] = P.NCPR
    IDR_data['FCR'] = P.FCR
    IDR_data['kappa'] = P.kappa
    IDR_data['hy_KD'] = P.hydrophobicity
    IDR_data['hy_clust'] = P.compute_kappa_x(["I","L","V","A","M"])
    IDR_data['aro_clust'] = P.compute_kappa_x(["F","Y","T","W","L"])
    IDR_data['AEM_clust'] = P.compute_kappa_x(["F","Y","L",'W'],['D','E'])
    IDR_data['SCD'] = P.SCD
    IDR_data['SHD'] = P.SHD
    IDR_data['NCPR'] = P.complexity
#     P=AnalyticalFRC(IDR_seq)
#     afrc_Rg = P.get_mean_radius_of_gyration()
#     afrc_Re = P.get_mean_end_to_end_distance()
#     #compile data for this IDR


#     IDRIDR_data['afrc_Re'] = afrc_Re
#     IDR_data['afrc_Rg'] = afrc_Rg
#     IDR_data['norm_Re'] = Re/afrc_Re
#     IDR_data['norm_Rg'] = Rg/afrc_Rg

    # Normalize Re and Rg (assuming afrc_Re and afrc_Rg are calculated and not zero)
    # Uncomment and use these lines if you have the analytical FRC calculations ready
    # IDR_data['norm_Re'] = IDR_data['Re'] / IDR_data['afrc_Re'] if 'afrc_Re' in IDR_data and IDR_data['afrc_Re'] else None
    # IDR_data['norm_Rg'] = IDR_data['Rg'] / IDR_data['afrc_Rg'] if 'afrc_Rg' in IDR_data and IDR_data['afrc_Rg'] else None
    
    # Append the data for this sequence to the results DataFrame
    # Convert IDR_data dictionary to a DataFrame with a single row
    single_idr_df = pd.DataFrame([IDR_data])
    # Append this row to the results_df
    results_df = pd.concat([results_df, single_idr_df], ignore_index=True)

# At this point, results_df contains all the calculated data for each sequence in 'ADseq', including the sequence itself
results_df
# Assuming 'df' and 'results_df' are your DataFrames
# And 'ADseq' is the column you want to match on
combined_df = pd.merge(df, results_df, on='ADseq', how='left')
```

```python
print(combined_df.columns)
filtered_df = combined_df[combined_df['CITED2']]
# Assuming 'df' is your DataFrame
filtered_df.to_csv('filtered_df.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame for rows where 'Hif1alpha' is True
filtered_df = combined_df[combined_df['CITED2']]

# Get values for 'Re_normalized' based on the 'status' condition
activating_values = filtered_df[filtered_df['status'] == 'activating']['FCR']
deactivating_values = filtered_df[filtered_df['status'] == 'deactivating']['FCR']

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(20, 15))

# Create boxplots
boxplots = ax.boxplot([activating_values, deactivating_values], labels=['Activating', 'Deactivating'], patch_artist=False)
# Setting the linewidth for the boxplot components

for element in ['boxes', 'whiskers', 'caps', 'medians']:
    plt.setp(boxplots[element], linewidth=5,color='black')  # Increase the linewidth here
# Add scatter plot with fixed x positions for each category
x_activating = [1] * len(activating_values)  # X position for all 'activating' points
x_deactivating = [2] * len(deactivating_values)  # X position for all 'deactivating' points

# Plotting the scatter points
ax.scatter(x_activating, activating_values, color='blue',s=700, alpha=0.25)
ax.scatter(x_deactivating, deactivating_values, color='red',alpha=0.25,s=700)

# Setting labels for axes and title
ax.set_ylabel('Re normalized')
ax.set_title('Boxplots with Scatter for Activating and Deactivating Statuses')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# List of features to plot
features = ['Re_normalized', 'aspher', 'NCPR', 'FCR', 'kappa', 'hydrophobicity', 'hy_clust', 'aro_clust', 'AEM_clust', 'SCD', 'SHD']

# Determine the number of rows needed for 4 plots per row
n_rows = (len(features) + 3) // 4  # Ensures there are enough rows for all features

# Create a figure with subplots - one for each feature, 4 per row
fig, axs = plt.subplots(n_rows, 4, figsize=(65, 15*n_rows), sharex=True)

# Flatten the axs array for easier iteration if it's 2D (necessary when n_rows > 1)
axs = axs.ravel()

# Loop through each feature and create a subplot
for i, feature in enumerate(features):
    # Filter the DataFrame for rows where 'Hif1alpha' is True
    filtered_df = combined_df[combined_df['Hif1alpha']]
    
    # Get values for the current feature based on the 'status' condition
    activating_values = filtered_df[filtered_df['status'] == 'activating'][feature]
    deactivating_values = filtered_df[filtered_df['status'] == 'deactivating'][feature]
    
    # Create boxplots on the current axis
    ax = axs[i]
    boxplots = ax.boxplot([activating_values, deactivating_values], labels=['Activating', 'Deactivating'], patch_artist=False)
    # Setting the linewidth for the boxplot components
    for element in ['boxes', 'whiskers', 'caps', 'medians']:
        plt.setp(boxplots[element], linewidth=5, color='black')
        
    # Add scatter plot with fixed x positions for each category
    x_activating = [1] * len(activating_values)
    x_deactivating = [2] * len(deactivating_values)
    ax.scatter(x_activating, activating_values, color='blue', s=700,edgecolor='black',linewidths=5)
    ax.scatter(x_deactivating, deactivating_values, color='red', s=700,edgecolor='black',linewidths=5)
    
    # Setting labels for axes and title
    ax.set_ylabel(feature)
    ax.set_title(f'{feature}')
# Hide any unused subplots
for j in range(i + 1, len(axs)):
    axs[j].set_visible(False)  # Hide unused subplot
# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
plt.savefig('filename.png', dpi=1200)
```

```python
import matplotlib.pyplot as plt
import math

# Define the categories to be plotted on the y-axis
categories = ['aspher', 'NCPR', 'FCR', 'kappa', 'hy_KD', 'hy_clust','aro_clust', 'AEM_clust']

# Calculate the necessary layout dimensions
num_categories = len(categories)
plots_per_category = 2  # We have two plots per category: Hif1alpha and CITED2
max_rows = 4

# Ensure we have a multiple of plots_per_category per row
columns_per_row = max(plots_per_category, math.ceil((num_categories * plots_per_category) / max_rows))
if (columns_per_row % plots_per_category) != 0:
    columns_per_row += plots_per_category - (columns_per_row % plots_per_category)

# Calculate the total number of subplots needed and ensure it's a multiple of plots_per_category
total_plots = num_categories * plots_per_category
num_rows = math.ceil(total_plots / columns_per_row)

# Create the subplot grid
fig, axs = plt.subplots(num_rows, columns_per_row, figsize=(15, 3.1 * num_rows))
axs_flat = axs.flatten()

# Plot each category
for i, category in enumerate(categories):
    # Hif1alpha plot
    axs_flat[i * plots_per_category].scatter(filtered_df.loc[filtered_df['Hif1alpha'] == True, 'activity'],
                                             filtered_df.loc[filtered_df['Hif1alpha'] == True, category],
                                             alpha=0.5)
    axs_flat[i * plots_per_category].set_title(f'Hif1alpha - {category}')
    axs_flat[i * plots_per_category].set_xlabel('Activity')
    axs_flat[i * plots_per_category].set_ylabel(category)
    
    # CITED2 plot
    axs_flat[i * plots_per_category + 1].scatter(filtered_df.loc[filtered_df['CITED2'] == True, 'activity'],
                                                 filtered_df.loc[filtered_df['CITED2'] == True, category],
                                                 alpha=0.5, color='orange')
    axs_flat[i * plots_per_category + 1].set_title(f'CITED2 - {category}')
    axs_flat[i * plots_per_category + 1].set_xlabel('Activity')
    axs_flat[i * plots_per_category + 1].set_ylabel(category)
    if (category=='AEM_clust')|(category=='kappa'):
        axs_flat[i * plots_per_category + 1].set_ylim(-0.02,0.5)
        axs_flat[i * plots_per_category].set_ylim(-0.02,0.5)

# Hide any unused axes
for ax in axs_flat[num_categories * plots_per_category:]:
    ax.set_visible(False)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
